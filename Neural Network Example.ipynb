```python
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from tqdm import tqdm
from sklearn.model_selection import train_test_split
```

```python
raw_data = pd.read_csv("wdbc.csv", header=None)

raw_data
```

```python
x = raw_data.iloc[:,2:32]

x
```

```python
x_normalized = (x - x.min()) / (x.max() - x.min())

x_normalized
```

```python
y = raw_data[1].replace(['B'], 0).replace(['M'], 1)

y = pd.get_dummies(y, prefix='y')

y
```

```python
df = x_normalized.copy()
df = pd.concat([df, y], axis=1)

df
```

```python
x_data = df.iloc[:,:30]

x_data
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data.values, y.values, test_size=0.33, random_state=4)
```

```python
class MyNeuralNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.h1 = nn.Linear(30, 25)
        self.h2 = nn.Linear(25, 10)
        self.output = nn.Linear(10, 2)
        self.activate = nn.Sigmoid()
        self.softmax = nn.Softmax(dim=-1)
        
    def forward(self, x):
        x = self.h1(x)
        x = self.activate(x)
        
        x = self.h2(x)
        x = self.activate(x)
        
        x = self.output(x)
        x = self.activate(x)
        
        y = self.softmax(x)
        
        return y
```

```python
model = MyNeuralNetwork()

model
```

```python
x_train_tensor = torch.tensor(x_train).float()
y_train_tensor = torch.tensor(y_train).float()

x_test_tensor = torch.tensor(x_test).float()
y_test_tensor = torch.tensor(y_test).float()
```

```python
from torch.utils.data import Dataset

class CustomDataset(Dataset):
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
        self.n_samples = len(x)
        
    def __getitem__(self, index):
        return self.x[index], self.y[index]
    
    def __len__(self):
        return self.n_samples
```

```python
training_ds = CustomDataset(x_train_tensor, y_train_tensor)
```

```python
from torch.utils.data import DataLoader

batch_size = 5

train_loader = DataLoader(training_ds, batch_size=batch_size, shuffle=True, drop_last=False)
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

def train_fn():
    loop = tqdm(train_loader)
    
    ave_loss = 0
    count = 0
    
    for batch_idx, (data, targets) in enumerate(loop):
        # Forward
        predictions = model.forward(data)
        
        loss = loss_fn(predictions, targets)
        
        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        # Update tqdm
        loop.set_postfix(loss=loss.item())
        
        count += 1
        ave_loss += loss.item()
    
    ave_loss = ave_loss / count
    
    return ave_loss
```

```python
epochs = 100

for epoch in range(epochs):
    print("Epoch: {}".format(epoch))
    
    ave_loss = train_fn()
    
    print("Ave Loss: {}".format(ave_loss))
```

```python
import numpy as np

predictions = model.forward(x_test_tensor)

predictions = predictions.detach().cpu().numpy()
predictions = [[1,0] if _y[0] > _y[1] else [0,1] for _y in predictions]
predictions = np.array(predictions)
predictions
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix

y_true = [1 if _y[0] == 1 else 0 for _y in y_test]
y_pred = [1 if _y[0] == 1 else 0 for _y in predictions]

tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()

print("True Negative: {}".format(tn))
print("False Positive: {}".format(fp))
print("False Negative: {}".format(fn))
print("True Positive: {}".format(tp))
```

```python
precision = tp / (tp + fp)
recall = tp / (tp + fn)

print("Precision: {}".format(precision))
print("Recall: {}".format(recall))
```
